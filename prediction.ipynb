```python
import cv2 as cv
import numpy as np
import tensorflow as tf
```

```python
def preprocess_with_opencv(image_path):
    img=cv.imread(image_path)

    if img is None:
        raise FileNotFoundError(f'could not find image at path:{image_path}')
    
    img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    img=cv.resize(img,(214,214),interpolation=cv.INTER_AREA)
    img=img.astype(np.float32) / 255

    img=np.expand_dims(img,axis=0)

    return img
```

```python
model=tf.keras.models.load_model(r'models\NSFW_predictor.keras')
```

```python
test_img_path=r"D:\Udemy\NSFW detection\nsfw_project_dataset\train\sfw\0a5e0d5d349fdf368afdd2b1a6e668c5899625cc4db3565ae5fe1b93340c4fc6.jpg"
processed_img=preprocess_with_opencv(test_img_path)
prediction_raw=model.predict(processed_img)
sfw_score=prediction_raw[0][0]
```

```python
print(f"Probability Score: {sfw_score:.4f}")
if sfw_score >= 0.5:
    print(f"🚨 Verdict: NSFW Content Detected ({sfw_score * 100:.2f}%)")
else:
    print(f"✅ Verdict: Safe Content ({(1 - sfw_score) * 100:.2f}%)")
```
