```python
## function to make master list from given dataset
import os
folder_path='nsfw_dataset'
os.makedirs('sfw_masterclass',exist_ok=True)
os.makedirs('nsfw_masterclass',exist_ok=True)

def create_masterclass_dataset(folder_path):

    with open('sfw_masterclass/sfw_list.txt','w') as sw_file,\
        open('nsfw_masterclass/nsfw_list.txt','w') as nsw_file:
        for current_folder,subfolders,files in os.walk(folder_path):
            print('start transferring files from {}'.format(current_folder))
            for file_name in files:
                full_path=os.path.join(current_folder,file_name)
                category=current_folder.split('\\')[1]

                try:
                    if category in ['drawings','neutral']:
                        sw_file.write(full_path+'\n')                        
                    elif category in ['hentai','porn']:
                        nsw_file.write(full_path+'\n')
                    else:
                        pass
                except:
                    print('error transfering {}'.format(file_name))
```

```python
create_masterclass_dataset(folder_path)
```

```python
## shuffling and slicing master lists
import random
import shutil

def shuffling_slicing(path):
    with open(path,'r') as file:
        file_lines=file.read().splitlines()
    
    split_point=int(len(file_lines)*0.80)
    
    ## shuffling of files
    random.shuffle(file_lines)

    train_list=file_lines[:split_point]
    val_list=file_lines[split_point:]

    return train_list,val_list

def save_sfw_data(sfw_path):
    training,val=shuffling_slicing(sfw_path)
    try:
        print('start transfering SFW data.....')
        for img_path in training:
            shutil.copy(src=img_path,dst=r'nsfw_project_dataset\train\sfw')
        for img_path in val:
            shutil.copy(src=img_path,dst=r'nsfw_project_dataset\val\sfw')
    except Exception as e:
        print(f'error occured in transferring SFW data: {e}')

def save_nsfw_data(nsfw_path):
    training,val=shuffling_slicing(nsfw_path)
    try:
        print('start transfering NSFW data.....')
        for img_path in training:
            shutil.copy(src=img_path,dst=r'nsfw_project_dataset\train\nsfw')
        for img_path in val:
            shutil.copy(src=img_path,dst=r'nsfw_project_dataset\val\nsfw')
    except Exception as e:
        print(f'error occured in transferring NSFW data: {e}')
```

```python
save_sfw_data(r'sfw_masterclass\sfw_list.txt')
save_nsfw_data(r'nsfw_masterclass\nsfw_list.txt')
```
